```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

model_id = "Qwen/Qwen2.5-Coder-7B-Instruct"  # NOTE: Qwen-2.5 8B Coder is not public yet; use 7B chat or Qwen1.5-7B for now

tokenizer = AutoTokenizer.from_pretrained(model_id, trust_remote_code=True)
model = AutoModelForCausalLM.from_pretrained(model_id, device_map="auto", torch_dtype=torch.float16, trust_remote_code=True)
model.eval()
```

```python
def build_react_prompt(task):
    return f"""You are an AI smart contract assistant using ReAct.
Your job is to help generate and iteratively refine OpenZeppelin smart contracts using the Wizard API.

User Goal:
{task}

Think step-by-step. Use the following format:

Thought: Describe what to do
Action: Provide the change or setting
Observation: Result of the action
...

End your response with FINAL ANSWER: <the completed options or JSON>

Begin:

User: {task}
"""
```

```python
def query_model(prompt, max_new_tokens=512):
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    with torch.no_grad():
        output = model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            do_sample=True,
            temperature=0.7,
            top_p=0.95,
            eos_token_id=tokenizer.eos_token_id
        )
    return tokenizer.decode(output[0], skip_special_tokens=True)
```
